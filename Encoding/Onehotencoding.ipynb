```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

Why used onehot encoding :


when categorical data having norminal value it's means it contain multiple category value look below example its present multiple car and type

```python
df['owner'].value_counts()
```

1. OneHotEncoding using Pandas

```python
pd.get_dummies(df,columns=['fuel','owner'],dtype=int)
```

2. K-1 OneHotEncoding

    when multiple column having multiple value then need to drop one column known as multiconality



```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True,dtype=int)
```

3. OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=True,dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_train_new
```

```python
X_test_new= ohe.transform(X_test[['fuel','owner']])
X_test_new
```


4. OneHotEncoding with Top Categories


To take only import column and other column should be others


```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype=int).sample(5)
```
